## Fixing the dates so they display better on the graphs
This not only makes the date axis more readable but also spaces the data out properly on a time scale and plots data from different series over each other properly.

```python
import datetime as dt
```

```python
"""
just a note: there is a bug in pandas where passing in the index as an arg when creating a dataframe
is not as robust as assinging the index attribute to something after the dataframe has been created
sometimes all the data become NaN, see stack overflow link for more info
https://stackoverflow.com/questions/29706825/pandas-version-0-16-0-after-changing-dataframe-index-all-values-become-nan
"""


"""
set_timestamp_to_index takes in a pandas DataFrame, df, and sets its index to the the timestamp of each row
pandas plots time series really nicely if the index has the time stored as datetime.datetime objects

timestamp_local is the column where the timestamps are stored in YYYY-MM-DDTHH:MM:SSZ format

return: no return needed since the df is passed by reference
"""
def set_timestamp_to_index(df):
    index = []
    for ts in df.timestamp_local:
        Y = int(ts[:4])
        M = int(ts[5:7])
        D = int(ts[8:10])
        H = int(ts[11:13])
        T = int(ts[14:16])
        S = int(ts[17:19])
        index.append(dt.datetime(Y, M, D, H, T, S))
    df.index = index
```

```python
for SN in SN_list:
    set_timestamp_to_index(SN)
```

```python
for SN in SN_list:
    SN.pm1.plot()
```

## Characterize data with histograms

```python
def histograms(df, temp_box=False, temp_manifold=False, rh_manifold=False, pressure=False,
               noise=False, solar=False, wind_dir=False, wind_speed=False, co2=False, voc=False,
               co=False, no=False, no2=False, o3=False, pm1=False, pm25=False, pm10=False):
    bins = 100
    if(temp_box):
        df.hist(column='temp_box', bins=bins)
    if(temp_manifold):
        df.hist(column='temp_manifold', bins=bins)
    if(rh_manifold):
        df.hist(column='rh_manifold', bins=bins)
    if(pressure):
        df.hist(column='pressure', bins=bins)
    if(noise):
        df.hist(column='noise', bins=bins)
    if(solar):
        df.hist(column='solar', bins=bins)
    if(wind_dir):
        df.hist(column='wind_dir', bins=bins)
    if(wind_speed):
        df.hist(column='wind_speed', bins=bins)
    if(co2):
        df.hist(column='co2', bins=bins)
    if(voc):
        df.hist(column='voc', bins=bins)
    if(co):
        df.hist(column='co', bins=bins)
    if(no):
        df.hist(column='no', bins=bins)
    if(no2):
        df.hist(column='no2', bins=bins)
    if(o3):
        df.hist(column='o3', bins=bins)
    if(pm1):
        df.hist(column='pm1', bins=bins)
    if(pm25):
        df.hist(column='pm25', bins=bins)
    if(pm10):
        df.hist(column='pm10', bins=bins)
```

```python
for SN in SN_list:
    # it appears that data only exists for pm1, pm25, and pm10
    # histograms(SN, co2=True, co=True, no=True, no2=True, o3=True, pm1=True, pm25=True, pm10=True)
    print(SN.sn[0])
    histograms(SN, pm1=True, pm25=True, pm10=True)
```
